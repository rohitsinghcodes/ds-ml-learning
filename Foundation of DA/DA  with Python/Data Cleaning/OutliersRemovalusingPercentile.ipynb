Import Required Lib

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

Load The DataSet

```python
# Load the dataset
df = pd.read_csv('data.csv')
```

Visualizing The Data

```python
# Visualize data distribution
sns.boxplot(x=df['Scores'])
plt.show()
```

Calculate the Percentiles

```python
# # Calculate the lower and upper percentiles
# df['Scores'] = pd.to_numeric(df['scores'], errors="coerce")
max_limit = df["Scores"].quantile(0.8)
min_limit = df["Scores"].quantile(0.2)

# Print the percentile values
print("Lower Percentile (2nd):", min_limit)
print("Upper Percentile (99th):", max_limit)
```

Identify and Remove Outliers

```python
filtered_df = df[(df['Scores'] >= min_limit) & (df['Scores'] <= max_limit)]

# Display the shape of the original and filtered dataset
print(f"Original dataset shape: {df.shape}")
print(f"Filtered dataset shape: {filtered_df.shape}")
```

Verify the Results

```python
# Visualize the cleaned data distribution
sns.boxplot(x=filtered_df['Scores'])
plt.show()
```
